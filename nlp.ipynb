```python

#NER
import nltk

# Download required packages
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

# Input from user
text = input("Enter sentence: ")# steve jobs  founded apple

# Tokenize words
words = nltk.word_tokenize(text)

# POS tagging
tags = nltk.pos_tag(words)

# Named Entity Recognition
ner = nltk.ne_chunk(tags)

# Output
print("\nNamed Entities:\n")
print(ner)
```

```python
#morpho_analysis
import spacy
nlp = spacy.load("en_core_web_sm")
sentence = input("Enter a Sentence: ")#  The company started in 2010 and now it is a global brand

doc = nlp(sentence)
print("{:<15} {:<15} {:<10} {:<10} {:<10} {:<10}".format(
    "Word","Lemma","POS","Person","Tense","Number"
))
print("-"*70)
for token in doc:
  person = token.morph.get("Person")
  tense = token.morph.get("Tense")
  number = token.morph.get("Number")

  person = person[0] if person else "-"
  tense = tense[0] if tense else "-"
  number = number[0] if number else "-"

  print("{:<15} {:<15} {:<10} {:<10} {:<10} {:<10}".format(
      token.text,
      token.lemma_,
      token.pos_,
      person,
      tense,
      number
  ))
```

```python
#Edit_distance
import java.util.Scanner;

public class MinimumEditDistance {

    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Enter source string: ");
        String source = sc.nextLine();

        System.out.print("Enter target string: ");
        String target = sc.nextLine();

        int m = source.length();
        int n = target.length();

        int[][] dp = new int[m + 1][n + 1];


        for (int i = 0; i <= m; i++) {
            dp[i][0] = i;
        }

        for (int j = 0; j <= n; j++) {
            dp[0][j] = j;
        }

        for (int i = 1; i <= m; i++) {
            for (int j = 1; j <= n; j++) {

                if (source.charAt(i - 1) == target.charAt(j - 1)) {
                    dp[i][j] = dp[i - 1][j - 1];
                } else {
                    dp[i][j] = 1 + Math.min(
                            dp[i - 1][j],
                            Math.min(
                                    dp[i][j - 1],
                                    dp[i - 1][j - 1]
                            )
                    );
                }
            }
        }

        System.out.println("\nEdit Distance Table:");
        for (int i = 0; i <= m; i++) {
            for (int j = 0; j <= n; j++) {
                System.out.print(dp[i][j] + " ");
            }
            System.out.println();
        }

        System.out.println("\nMinimum Edit Distance: " + dp[m][n]);
    }
}
```

```python
#POS_viterbi

import nltk
from collections import defaultdict

train_data = [
    [("I", "PRON"), ("eat", "VERB"), ("rice", "NOUN")],
    [("She", "PRON"), ("likes", "VERB"), ("apple", "NOUN")]
]

tag_counts = defaultdict(int)
transition_prob = defaultdict(dict)
emission_prob = defaultdict(dict)

for sent in train_data:

    prev_tag = None

    for word, tag in sent:

        tag_counts[tag] += 1

        emission_prob[tag][word.lower()] = \
            emission_prob[tag].get(word.lower(), 0) + 1

        if prev_tag:
            transition_prob[prev_tag][tag] = \
                transition_prob[prev_tag].get(tag, 0) + 1

        prev_tag = tag

tags = list(tag_counts.keys())

def viterbi(sentence):

    words = sentence.lower().split()

    result = []

    for word in words:

        best_tag = None
        best_prob = 0

        for tag in tags:

            prob = emission_prob[tag].get(word, 0)

            if prob > best_prob:
                best_prob = prob
                best_tag = tag

        result.append((word, best_tag))

    return result

sentence = input("Enter a sentence: ")

print(viterbi(sentence))
```

```python
#POS_hmm
import nltk
from nltk.corpus import treebank
from collections import defaultdict, Counter

nltk.download('treebank')

tagged_sentences = treebank.tagged_sents()

transition_counts = defaultdict(Counter)
emission_counts = defaultdict(Counter)
tag_counts = Counter()

for sentence in tagged_sentences:

    prev_tag = None

    for word, tag in sentence:

        tag_counts[tag] += 1
        emission_counts[tag][word.lower()] += 1

        if prev_tag is not None:
            transition_counts[prev_tag][tag] += 1

        prev_tag = tag

transition_prob = defaultdict(dict)
emission_prob = defaultdict(dict)

for prev_tag in transition_counts:

    total = sum(transition_counts[prev_tag].values())

    for tag in transition_counts[prev_tag]:

        transition_prob[prev_tag][tag] = \
            transition_counts[prev_tag][tag] / total

for tag in emission_counts:

    total = sum(emission_counts[tag].values())

    for word in emission_counts[tag]:

        emission_prob[tag][word] = \
            emission_counts[tag][word] / total

def hmm_greedy_tagging(sentence):

    words = sentence.lower().split()

    tags_list = list(tag_counts.keys())

    predicted_tags = []

    prev_tag = None

    for word in words:

        best_tag = None
        max_prob = 0

        for tag in tags_list:

            emis = emission_prob[tag].get(word, 1e-6)

            if prev_tag is None:
                prob = emis
            else:
                trans = transition_prob[prev_tag].get(tag, 1e-6)
                prob = emis * trans

            if prob > max_prob:
                max_prob = prob
                best_tag = tag

        predicted_tags.append((word, best_tag))

        prev_tag = best_tag

    return predicted_tags

sentence = input("Enter a sentence: ")

print(hmm_greedy_tagging(sentence))
```

```python
#dfa_without_2_1s
import java.util.Scanner;

public class DFA {

    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Enter binary string: "); #10101 #11101
        String input = sc.nextLine();

        int state = 0;

        for (int i = 0; i < input.length(); i++) {

            char ch = input.charAt(i);

            if (ch != '0' && ch != '1') {
                System.out.println("Invalid input!");
                return;
            }

            switch (state) {

                case 0:

                    if (ch == '0')
                        state = 0;
                    else
                        state = 1;

                    break;

                case 1:

                    if (ch == '0')
                        state = 0;
                    else
                        state = 2;

                    break;

                case 2:

                    state = 2;
                    break;
            }
        }

        if (state == 0 || state == 1)
            System.out.println("Accepted.No consective 1's");
        else
            System.out.println("Rejected.It has consective 1's");

        sc.close();
    }
}
```

```python
#DFA_TT

import java.util.Scanner;

public class DFA_TT {

    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Enter number of states: ");
        int n = sc.nextInt();

        System.out.print("Enter number of input symbols: ");
        int m = sc.nextInt();

        int[][] table = new int[n][m];

        System.out.println("Enter transition table:");

        for (int i = 0; i < n; i++) {

            for (int j = 0; j < m; j++) {

                System.out.print("δ(q" + i + ", " + j + ") = ");
                table[i][j] = sc.nextInt();
            }
        }

        System.out.print("Enter start state: ");
        int start = sc.nextInt();

        System.out.print("Enter number of final states: ");
        int f = sc.nextInt();

        int[] finalStates = new int[f];

        System.out.println("Enter final states:");

        for (int i = 0; i < f; i++) {

            finalStates[i] = sc.nextInt();
        }

        sc.nextLine();

        System.out.print("Enter input string: ");
        String str = sc.nextLine();

        int state = start;

        for (int i = 0; i < str.length(); i++) {

            int input = str.charAt(i) - '0';

            state = table[state][input];
        }

        boolean accept = false;

        for (int i = 0; i < f; i++) {

            if (state == finalStates[i]) {

                accept = true;
                break;
            }
        }

        if (accept)
            System.out.println("Result: Accepted");
        else
            System.out.println("Result: Rejected");

        sc.close();
    }
}
```
