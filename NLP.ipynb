```python
#morpho_analysis
import spacy
nlp = spacy.load("en_core_web_sm")
sentence = input("Enter a Sentence: ") #The company started in 2010 and now it is a global brand
```

```python
doc = nlp(sentence)
print("{:<15} {:<15} {:<10} {:<10} {:<10} {:<10}".format(
    "Word","Lemma","POS","Person","Tense","Number"
))
print("-"*70)
for token in doc:
  person = token.morph.get("Person")
  tense = token.morph.get("Tense")
  number = token.morph.get("Number")

  person = person[0] if person else "-"
  tense = tense[0] if tense else "-"
  number = number[0] if number else "-"

  print("{:<15} {:<15} {:<10} {:<10} {:<10} {:<10}".format(
      token.text,
      token.lemma_,
      token.pos_,
      person,
      tense,
      number
  ))
```

```python
import nltk
import string
from nltk.corpus import brown, stopwords, genesis

# 1. Download necessary NLTK corpora
nltk.download('brown')
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('genesis')

# 2. Load the Genesis corpus and extract words and sentences
gen_words = genesis.words()
gen_sentences = genesis.sents()

print("--- Corpus Statistics ---")
print("Sample from Genesis Corpus:", gen_words[:10])

# Total words including punctuation
total_words = len(gen_words)
print("Total words (including punctuation):", total_words)

# Total words excluding punctuation
no_punct_words = [w for w in gen_words if w.isalpha()]
print("Total words (excluding punctuation):", len(no_punct_words))
print("Total sentences:", len(gen_sentences))

# 3. Remove stop words
stop_words = set(stopwords.words('english'))
filtered_words = []
for w in no_punct_words:
    if w.lower() not in stop_words:
        filtered_words.append(w.lower())

print("\n--- Filtered Words (Top 10) ---")
print(filtered_words[:10])

# 4. Create sorted vocabulary and calculate word frequency
vocabulary = sorted(set(filtered_words))
print(f"\nUnique words in vocabulary: {len(vocabulary)}")

word_freq = {}
for w in filtered_words:
    if w in word_freq:
        word_freq[w] += 1
    else:
        word_freq[w] = 1

# Display frequency of specific sample words
print("Sample Frequencies:")
for i in range(5):
    word = filtered_words[i]
    print(f"{word}: {word_freq[word]}", end=", ")
print()

# 5. Display vocabulary samples by starting letter
print("\n--- Vocabulary Sample by Alphabet ---")
for letter in string.ascii_lowercase:
    letter_words = [w for w in vocabulary if w.startswith(letter)]
    if letter_words:
        print(f"{letter.upper()}: {letter_words[:3]}")

# 6. Define slang dictionary
slang = {
    "bro": "Brother",
    "ngl": "not gonna lie",
    "idk": "I don't know",
    "abt": "about",
    "nlp": "Natural Language Processing"
}

# 7. Replace slang and expand sentence
sentence = "Bro ngl, idk anything abt nlp."
expanded = []

for w in sentence.split():
    # Remove punctuation for matching
    clean_word = w.lower().strip(".,")
    if clean_word in slang:
        expanded.append(slang[clean_word])
    else:
        expanded.append(w)

# 8. Display results
print("\n--- Slang Expansion ---")
print("Original sentence:", sentence)
print("Expanded sentence:", " ".join(expanded))
```

```python
#ngrams
import nltk
from nltk.util import ngrams
from collections import Counter

# 1. Download required NLTK resources [cite: 22-23]
nltk.download('punkt')
nltk.download('punkt_tab')

# 2. Get the input text corpora [cite: 32-41, 43-47]
print("Enter sentences one by one. Type 'exit' to stop:\n")
sentences = []
while True:
    line = input()
    if line.lower() == "exit":
        break
    sentences.append(line)

text = " ".join(sentences)
tokens = text.lower().split()
print("Tokens:", tokens) # [cite: 48]

# 3. Generate sequences and compute frequency counts [cite: 50-52]
unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

unigram_freq = Counter(unigrams)
bigram_freq = Counter(bigrams)
trigram_freq = Counter(trigrams)

# 4. Calculate probabilities [cite: 86-97]
total_unigrams = sum(unigram_freq.values())
bigram_prob = {bg: count / unigram_freq[(bg[0],)] for bg, count in bigram_freq.items()}
trigram_prob = {tg: count / bigram_freq[(tg[0], tg[1])] for tg, count in trigram_freq.items()}

# 5. Analysis Functions [cite: 98-127]
def analyze_unigram(unigram_freq, total):
    data = [(u[0], f, f/total) for u, f in unigram_freq.items()]
    return sorted(data, key=lambda x: x[1], reverse=True), "Unigram"

def analyze_bigram(words, b_prob, b_freq):
    last = words[-1]
    data = [(bg[1], b_freq[bg], p) for bg, p in b_prob.items() if bg[0] == last]
    return data, "Bigram"

def analyze_trigram(words, t_prob, t_freq):
    last_bg = (words[-2], words[-1])
    data = [(tg[2], t_freq[tg], p) for tg, p in t_prob.items() if (tg[0], tg[1]) == last_bg]
    return data, "Trigram"

# 6. Interactive Prediction with specific output formatting [cite: 129-154]
while True:
    user_input = input("\nEnter a phrase (or type 'exit'): ")
    if user_input.lower() == "exit":
        break

    words = nltk.word_tokenize(user_input.lower())
    analysis_data, model_used = [], ""

    if len(words) >= 2:
        analysis_data, model_used = analyze_trigram(words, trigram_prob, trigram_freq)

    if not analysis_data and len(words) >= 1:
        analysis_data, model_used = analyze_bigram(words, bigram_prob, bigram_freq)

    if not analysis_data:
        analysis_data, model_used = analyze_unigram(unigram_freq, total_unigrams)

    # Formatted Output Table [cite: 146-150, 157]
    print(f"--- {model_used} ---")
    print(f"{'Next Word':<15} | {'Frequency':<10} | {'Probability'}")
    print("-" * 45)
    for word, freq, prob in analysis_data:
        print(f"{word:<15} | {freq:<10} | {prob:.4f}")
#I like nlp
#I like ai
#I enjoy ml
#exit
```
